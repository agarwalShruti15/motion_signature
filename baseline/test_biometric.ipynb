# plot confusion matrices

```python


# is_trg is True then use the face identity, else use the behaviour identity
def get_acc_FF(in_results, is_trg=True):
    
    dbs = ['FF_orig', 'FF_Deepfakes', 'FF_FaceSwap', 'FF_Face2Face', 'FF_NeuralTextures']
    acc = {}
    for d in dbs:
        lbl_dict = get_FF_src2behav_dict('/data/home/shruti/voxceleb/vgg/leaders/', d)
        true_lbl = np.array(in_results[d]['actual lbl'])
        pred_lbl = np.array(in_results[d]['pred lbl'])
        if not is_trg:
            pred_lbl = np.array([lbl_dict[f] for f in pred_lbl])
        acc[d] = len(np.argwhere(true_lbl == pred_lbl))/len(true_lbl)
    
    return acc

def confusion_matrix(in_result_leader):
    
    lbl_dict = get_leaders_lblToName_dict()
    out_confusion_arr = np.zeros((len(in_result_leader.keys()), len(lbl_dict.keys())))
    test_keys = list(in_result_leader.keys())
    
    for i in range(len(test_keys)):
        
        for j in in_result_leader[test_keys[i]]['pred lbl']:
            out_confusion_arr[i, j] = out_confusion_arr[i, j] + 1
        out_confusion_arr[i, :] = out_confusion_arr[i, :]/len(in_result_leader[test_keys[i]]['pred lbl'])
        
    return pd.DataFrame(data=out_confusion_arr, columns=[lbl_dict[f] for f in range(len(lbl_dict.keys()))], 
                index = test_keys)
```
